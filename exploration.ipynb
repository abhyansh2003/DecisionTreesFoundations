```python
# There's a columns 'Unnamed 32' which has null values so we've to drop it.
df = df.drop(['id', 'Unnamed: 32'], axis=1)
df.shape
```

```python
print("Unique values in the output labels: ")
df['diagnosis'].unique()
```

```python
sns.countplot(x='diagnosis', data=df)
plt.title("Target distribution (B=Benign, M=Malignant)")
plt.show()

print("\n")
print(df['diagnosis'].value_counts())
```

```python
# Univariate Analysis
df.drop('diagnosis', axis=1).hist(bins=20, figsize=(18,12))
plt.show()
```

```python
# Bivariate Analysis
sns.boxplot(x='diagnosis', y='radius_mean', data=df)
plt.show()
```

```python
plt.figure(figsize=(15,12))
sns.heatmap(df.corr(), cmap='coolwarm', annot=True)
plt.title("Feature Correlation Heatmap")
plt.show()
```
